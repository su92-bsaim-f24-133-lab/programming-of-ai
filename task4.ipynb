```python
# Dynamic N-Queens Problem

def print_board(board, n):
    horizontal_line = "+" + "-----+" * n
    for i in range(n):
        print(horizontal_line)
        for j in range(n):
            print("| ", board[i][j], end="  ")
        print("|")
    print(horizontal_line)


def is_safe(board, row, col, n):
    
    # Check column
    for i in range(row):
        if board[i][col] == "Q":
            return False
    
    # Check left diagonal
    i = row - 1
    j = col - 1
    while i >= 0 and j >= 0:
        if board[i][j] == "Q":
            return False
        i -= 1
        j -= 1
    
    # Check right diagonal
    i = row - 1
    j = col + 1
    while i >= 0 and j < n:
        if board[i][j] == "Q":
            return False
        i -= 1
        j += 1
    
    return True


def solve_nqueens(board, row, n):
    
    if row == n:
        print("\nSolution Found:\n")
        print_board(board, n)
        return True
    
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row][col] = "Q"
            
            if solve_nqueens(board, row + 1, n):
                return True
            
            board[row][col] = " "   # Backtrack
    
    return False


# 🔹 User Input
n = int(input("Enter value of N: "))

# Create empty board
board = [[" " for i in range(n)] for i in range(n)]

# Solve
if not solve_nqueens(board, 0, n):
    print("No Solution Exists")
```
