#  Implement the Water Jug Problem using Depth-First Search
# (DFS) and print the rules applied at each step

```python
# Water Jug Problem using DFS (Simple Version)

max_x = 4
max_y = 3
goal = 2

visited = []

def dfs(x, y):
    
    
    if (x, y) in visited:
        return
    
    
    visited.append((x, y))
    
    print("Current State:", x, y)
    
    # Goal check
    if x == goal or y == goal:
        print("Goal Reached!")
        return
    
    # 1. Fill X
    print("Rule: Fill X")
    dfs(max_x, y)
    
    # 2. Fill Y
    print("Rule: Fill Y")
    dfs(x, max_y)
    
    # 3. Empty X
    print("Rule: Empty X")
    dfs(0, y)
    
    # 4. Empty Y
    print("Rule: Empty Y")
    dfs(x, 0)
    
    # 5. Pour X -> Y
    print("Rule: Pour X to Y")
    pour = min(x, max_y - y)
    dfs(x - pour, y + pour)
    
    # 6. Pour Y -> X
    print("Rule: Pour Y to X")
    pour = min(y, max_x - x)
    dfs(x + pour, y - pour)


# Start from (0,0)
dfs(0, 0)
```
