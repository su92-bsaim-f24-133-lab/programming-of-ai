```python
#Menu
print("1.fill x")
print("2.fill y")
print("3.empty x")
print("4.empty y")
print("5.pour x y until x is empty")
print("6.pour y x until y is full")
print("7.pour x y until x is full")
print("8.pour y x until y is empty")
print("9.quit")

#Initial values
def water_jug():
    x = 0
    y = 0
    xcap = 4
    ycap = 3
    goal = 2

    while True:
        print(f"Current state: x={x}, y={y}")
        rule = int(input("Enter your choice (1-9): "))
        if rule == 2:
            y = ycap
        elif rule == 1:
            x = xcap
        elif rule == 3:
            x = 0
        elif rule == 4:
            y = 0
        elif rule == 5:
            transfer = min(x, ycap - y)
            x -= transfer
            y += transfer
        elif rule == 6:
            transfer = min(y, xcap - x)
            y -= transfer
            x += transfer
        elif rule == 7:
            transfer = min(x, ycap - y)
            x -= transfer
            y += transfer
        elif rule == 8:
            transfer = min(y, xcap - x)
            y -= transfer
            x += transfer
        elif rule == 9:
            print("Quitting the game.")
            break
        else:
            print("Invalid choice. Please try again.")
            continue

        if x == goal or y == goal:
            print(f"Goal achieved! Current state: x={x}, y={y}")
            break

# start the game
water_jug()
    

```
